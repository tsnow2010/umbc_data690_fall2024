```python
import pandas as pd
import plotly.express as px

pd.set_option("display.max_rows", None)

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

print(df.shape)
df.sample(3)
```

```python
# Gives multiple stats on two columns

df[['Population, total','Life expectancy at birth, total (years)']].describe()
```

```python
# Let's change the column names to shorten them.

# Make dictionary
column_names = {
    "Life expectancy at birth, total (years)":'Life Expectancy',
    "GDP per capita (current US$)":'GDP per Capita',
    "Population, total": 'Population'
}
# Pass dictionary to .rename(columns=dict)
df2 = df.rename(columns=column_names)
df2.sample(3)
```

```python
# Gives Categorical data stats
df2.describe(include="object")
```

```python
df2.describe(include="object").T
```

```python
# Gives numerical and categorical stats
df2.describe(include="all")
```

```python
def assign_quartile(gdp):
  if gdp < 1862.89:
    quartile = '1st Quartile'
  elif gdp > 1862.89 and gdp <= 6048.30:
    quartile = '2nd Quartile'
  elif gdp > 6048.30 and gdp <= 22137.31:
    quartile = '3rd Quartile'
  else:
    quartile = '4th Quartile'
  return quartile


print(assign_quartile(7738.878804))
```

```python
# Adds separate column, 'GDP Quartile', applying the defined function, assign_quartile()
df2['GDP Quartile'] =  df2['GDP per Capita'].apply(assign_quartile)
df2.sample(3)
```
