```python
# Step 2: Creates 'GDP Quartile' column and lists the quartile the row falls under.
# Next, creates 'Health and Wealth Status' column which categorizes a row if
# a country and a year falls under (4) different categories.

def assign_quartile(gdp, x, y, z):
  if gdp < x:
    quartile = '1st Quartile (<=25%)'
  elif gdp > x and gdp <= y:
    quartile = '2nd Quartile (25-50%)'
  elif gdp > y and gdp <= z:
    quartile = '3rd Quartile (50-75%)'
  else:
    quartile = '4th Quartile (>75%)'
  return quartile

df_2020['GDP Quartile'] =  df_2020['GDP per Capita'].apply(assign_quartile, args=(x, y, z,))
df_2020.sample(10)
```

```python
def find_status(row):
  if (row['Life Expectancy Quartile'] == "1st Quartile (<=25%)") and (row['GDP Quartile'] == "1st Quartile (<=25%)"):
    return 'Poor and unhealthy'
  elif (row['Life Expectancy Quartile'] == "4th Quartile (>75%)") and (row['GDP Quartile'] == "1st Quartile (<=25%)"):
    return 'Poor but healthy'
  elif (row['Life Expectancy Quartile'] == "4th Quartile (>75%)") and (row['GDP Quartile'] == "4th Quartile (>75%)"):
    return 'Rich and healthy'
  elif (row['Life Expectancy Quartile'] == "3rd Quartile (50-75%)") and (row['GDP Quartile'] == "4th Quartile (>75%)"):
    return 'Rich and somewhat healthy'
  elif (row['Life Expectancy Quartile'] == "2nd Quartile (25-50%)") and (row['GDP Quartile'] == "4th Quartile (>75%)"):
    return 'Rich and somewhat unhealthy'
  elif (row['Life Expectancy Quartile'] == "1st Quartile (<=25%)") and (row['GDP Quartile'] == "4th Quartile (>75%)"):
    return 'Rich but unhealthy'
  else:
    return 'Other'

df_2020['Health and Wealth Status'] = df_2020.apply(find_status,axis=1)
df_2020.sample(10)
```

Step 2:
* Make dataframes for each dataset.
* Edit the dataframes to allow merging.
  * The naming convention for the states is different between the datasets. We need to convert state names to their abbreviations in the GDP dataset.

```python
# Thanks to MS Copilot, it made the below function to convert state names to their abbreviations.
def state_to_abbreviation(state_name):
    state_name = state_name.strip()
    state_abbreviations = {
        'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
        'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
        'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
        'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
        'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO',
        'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ',
        'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH',
        'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
        'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT',
        'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
    }
    return state_abbreviations.get(state_name, state_name)
```

```python
# Below extracts data desired from df_states and puts in df_states_new.
df_states_new = pd.DataFrame({'State':df_states['STATE'], 'Year': df_states['YEAR'], 'Life Expectancy':df_states['YEARS']})
df_states_new.sample(5)
```

```python
# We focus only on year 2021 for life expectancy dataset.
df_states_2021 = df_states_new[df_states_new['Year'] == 2021]
df_states_2021.sample(5)
```

```python
# Converts state names into abbreviations in GDP dataset.
df_gdp['State'] = df_gdp['State'].apply(state_to_abbreviation)
df_gdp.tail(5)
```

```python
# Finally, we merge the (2) DataFrames.
df_states_lf_gdp = pd.merge(df_states_2021, df_gdp, on=['State'])
df_states_lf_gdp.sample(5)
```

```python
# Let's edit the new DataFrame a bit.

df_states_lf_gdp_new = df_states_lf_gdp.drop(columns='Year') # We don't care about the year for life expectancy.
df_states_lf_gdp_new = df_states_lf_gdp_new.rename(columns = {'Rank Q2':"GDP Rank in 2024 Q2"}) # Rename GDP column
# DC has ..... in one cell.  Below removes the row altogether.
df_states_lf_gdp_new['GDP Rank in 2024 Q2'][8] = None
df_states_lf_gdp_new.dropna(axis=0,inplace=True)
# GDP values are 'object' type.  This converts them into int.
df_states_lf_gdp_new['GDP Rank in 2024 Q2'] = df_states_lf_gdp_new['GDP Rank in 2024 Q2'].astype(int)

```

Step 3: Make a scatter plot with life expectancy as a function of GDP rank in the USA.

```python
fig3 = px.scatter(
    df_states_lf_gdp_new,
    x="GDP Rank in 2024 Q2",
    y="Life Expectancy",
    hover_name="State"
)

fig3.show()
```
