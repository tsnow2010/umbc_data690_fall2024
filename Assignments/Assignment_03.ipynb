## Assignment 03 Part 2

Process column "11. What is your long-term career goal?" to find out the most popular career goals of the class. User Python code to parse the responses and count occurences of words, then use your naked eyes to make determination. Write your observation/analysis using Markdown.

````python
career_goal_dist = df["11. What is your long-term career goal?"]

# Below function is a comprehensive version of str.strip().  It removes all punctuation in punct_list.

punct_list = [',','?','.','/','!']

def strip_all(word, punct_list, index):
  if index < len(punct_list)-1:
    return strip_all(word.strip(punct_list[index]).lower(), punct_list, index=index+1)
  else:
    return word.strip(punct_list[index])

# Below code parses responses from column 11 and makes a dictionary to store each parsed word and its frequency.

# Initialize dicts and lists.
word_freq_dict = {'Word':[], 'Freq':[]}
word_dict = {}
omit_word_list = ['become']

# Parse responses, count words, and filter out unnecessary words.
for response in career_goal_dist:
  response_words = response.split()
  for word in response_words:
    word = strip_all(word,punct_list,index=0)
    # Removes words that are 4 or less characters and from the omit_word_list.
    # Most words 4 or less characters were articles, helper verbs, etc.
    if word in omit_word_list or len(word) <= 4:
      continue
    if word not in word_dict:
      word_dict[word] = 1
    else:
      word_dict[word] += 1

# Converts the above dictionary into another dictionary that can be converted to a DataFrame with columns 'Word' and 'Freq.'
for word in word_dict:
  word_freq_dict['Word'].append(word)
  word_freq_dict['Freq'].append(word_dict[word])

# The below statement requires a dict param with the structure: dict = {'column1':[contents], 'column2':[contents]}
word_freq_dist = pd.DataFrame.from_dict(word_freq_dict)

# Makes bar graph.
bar_fig3 = px.bar(
    word_freq_dist,
    x='Word',
    y='Freq',
    color='Word',
    title='Word Frequencies in Career Goal Responses',
    text_auto=True,
    labels={'Freq':'Frequency'}
)

bar_fig3.update_layout(showlegend=False)

bar_fig3.show()
````
